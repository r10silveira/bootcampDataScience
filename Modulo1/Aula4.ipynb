```python
import pandas as pd
fonte = "../dados/A160324189_28_143_208.csv"

dados = pd.read_csv(fonte,encoding="latin1",skiprows=3,sep=";",decimal=",",skipfooter=12)
pd.options.display.float_format = "{:.2f}".format
dados.head()
```

```python
dados.mean()
```

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0,"Unidade da Federação")
usaveis = dados[colunas_usaveis]
usaveis = usaveis.set_index("Unidade da Federação")
usaveis = usaveis.drop("Total",axis=1)
usaveis.head()
```

```python
#Removendo as colunas antes de FEV/2008
dados_traspostos = usaveis.T[6:]
usaveis = dados_traspostos.T
usaveis.head()
```

```python
usaveis["Total"] = usaveis.sum(axis=1)
ordenado_por_total = usaveis.sort_values("Total",ascending=False)
ordenado_por_total = ordenado_por_total.drop("Total",axis=1)
ordenado_por_total.head()
```

```python
ordenado_por_total = ordenado_por_total / 1000000
ordenado_por_total.head()
```

```python
mes_recente = ordenado_por_total.columns[-1:]
gastos_mais_recente = ordenado_por_total[mes_recente]
gastos_mais_recente = gastos_mais_recente.sort_values("2020/Jul",ascending=False)
```
