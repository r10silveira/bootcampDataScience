```python
import pandas as pd
fonte = "../dados/A160324189_28_143_208.csv"

dados = pd.read_csv(fonte,encoding="latin1",skiprows=3,sep=";",decimal=",",skipfooter=12)
pd.options.display.float_format = "{:.2f}".format
dados.head()
```

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0,"Unidade da Federação")
usaveis = dados[colunas_usaveis]
usaveis = usaveis.set_index("Unidade da Federação")
#Removendo as colunas antes de FEV/2008
dados_traspostos = usaveis.T[6:]
usaveis = dados_traspostos.T
usaveis.head()
```

```python
ordenado_por_total = usaveis.sort_values("Total",ascending=False)
ordenado_por_total = ordenado_por_total.drop("Total",axis=1)
ordenado_por_total.head()
```

```python
ordenado_por_total = ordenado_por_total / 1000000
ordenado_por_total.head()
```

```python
ordenado_por_total = ordenado_por_total / 1000000
ordenado_por_total.head()
```

# Desafio 1 - Buscar na documentação do Matplotlib como colocar um grid nos gráficos e adicionar nos gráficos de barra.

```python
mes_recente = ordenado_por_total.columns[-1:]
gastos_mais_recente = ordenado_por_total[mes_recente]
gastos_mais_recente = gastos_mais_recente.sort_values("2020/Jul",ascending=False)
```

```python
import matplotlib.pyplot as plt
ax = gastos_mais_recente.plot(kind='bar',figsize=(10,6))
ax.set_title("Valores Gastos em Julho/2020")
plt.ylabel("Valores Gastos")
plt.xlabel("Estados")
plt.grid(True,axis="y",ls="--",linewidth=2)
```

# Desafio 3 - Fazer o cálculo proporcional a população do seu estado e mais um a sua escolha.

```python
# População de Minas Gerais em 2020 = 21.292.666
# População de São paulo em 2020 =  46.289.333

populacao_SP = 46289333
populacao_MG = 21292666

gastos_recente_populacao = usaveis[mes_recente]
gastos_recente_populacao = gastos_recente_populacao.query("`Unidade da Federação` == '35 São Paulo' | `Unidade da Federação` == '31 Minas Gerais'")
gastos_recente_populacao.loc[gastos_recente_populacao.index == "31 Minas Gerais","Populacao"] = populacao_MG
gastos_recente_populacao.loc[gastos_recente_populacao.index == "35 São Paulo","Populacao"] = populacao_SP
gastos_recente_populacao
```

```python
ax = (gastos_recente_populacao["2020/Jul"] / gastos_recente_populacao["Populacao"]).plot(kind="bar",figsize=(10,6),grid=True)

ax.set_title("Gastos com internações porporcional percaptita dos estados de Minas e São Paulo em Jul de 2020")
plt.ylabel("Valor Gasto percapita")
plt.xlabel("Estados")
plt.show()
```
