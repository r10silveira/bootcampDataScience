```
import pandas as pd
fonte = "https://raw.githubusercontent.com/alura-cursos/agendamento-hospitalar/main/dados/A160324189_28_143_208.csv"

dados = pd.read_csv(fonte,encoding="latin1",skiprows=3,sep=";",decimal=",",skipfooter=12)
pd.options.display.float_format = "{:.2f}".format

dados.head()



```

```
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0,"Unidade da Federação")
usaveis = dados[colunas_usaveis]
usaveis = usaveis.set_index("Unidade da Federação")
usaveis = usaveis.drop("Total",axis=1)
```

Desafio 01: Reposicionar a legenda do gráfico em uma posição mais adequada

```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = usaveis.T.plot(figsize=(10,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
ax.legend(loc=0, bbox_to_anchor=(1,1))


plt.show()
```

Desafio 02: Plotar o gráfico de linha com apenas 5 estados de sua preferência

```
cinco_estados = usaveis.T[["11 Rondônia","12 Acre","13 Amazonas","14 Roraima","15 Pará"]]

ax = cinco_estados.plot(figsize=(10,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))


plt.show()
```
