# Desafio 04 - Desenvolver uma função para a limpeza e tratamento dos dados, retornando duas tabelas "população" e "unida".

```
def ler_limpar_tabela_populacao_mesclar_gastos_recentes(dados_populacao,df_gastos_recentes):
    from io import StringIO

    dados_populacaoio = StringIO(dados_populacao)

    #remove dados núlos
    populacao = pd.read_csv(dados_populacaoio,sep="\t").dropna()

    #Cria coluna com nome populacao  e atribui o mesmo valor porém remove espaços entre os valores e converte para integer
    populacao["populacao"] = populacao["População"].str.replace(" ","").astype(int)

    #cria coluna com o intuito de renomear
    populacao["uf"] = populacao["Unidade federativa"]

    #Seleciona só as colunas que interessa, set indice e remove espaços do indice
    populacao = populacao[["uf","populacao"]]
    populacao = populacao.set_index("uf")
    populacao.index = populacao.index.str.strip()

    #Remove a numeração e espaço de todas os nome de estados
    if df_gastos_recentes.index.str[:2][0].isnumeric():
        df_gastos_recentes.index = df_gastos_recentes.index.str[3:]

    #remove nome de estados repetidos caso haja
    for estado in df_gastos_recentes.index:
        #print(f"{estado} {estado}")
        populacao.index = populacao.index.str.replace(f"{estado} {estado}",estado)
        gastos_populacao_recente = populacao.join(df_gastos_recentes)
    
    #retorna dataframe da população e retorna a datafame com a população e gastos mais recentes
    return populacao,gastos_populacao_recente
```

```
populacao, gastos_recentes_populacao = ler_limpar_tabela_populacao_mesclar_gastos_recentes(dados_populacao,gastos_do_mais_recente)

populacao
```
